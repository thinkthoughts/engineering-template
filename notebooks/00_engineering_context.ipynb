## 1. Load the repository in Colab or local Jupyter

This bootstrap cell uses the current repository when available and otherwise clones the repository into a fresh Colab workspace.

````python
from pathlib import Path
import os
import subprocess
import sys

REPOSITORY_URL = "https://github.com/thinkthoughts/benefits-distribution.git"
PACKAGE_NAME = "benefits_distribution"
CURRENT = Path.cwd().resolve()

ROOT = next(
    (
        candidate
        for candidate in (CURRENT, CURRENT.parent)
        if (candidate / "src" / PACKAGE_NAME).exists()
    ),
    None,
)

if ROOT is None:
    ROOT = Path("/content/benefits-distribution")
    if not ROOT.exists():
        subprocess.run(["git", "clone", REPOSITORY_URL, str(ROOT)], check=True)

SRC = ROOT / "src"
PACKAGE_DIR = SRC / PACKAGE_NAME

if not PACKAGE_DIR.exists():
    raise RuntimeError(f"Expected Python package at {PACKAGE_DIR}")

if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

os.chdir(ROOT)

print("Repository root:", ROOT)
print("Python source:", SRC)
print("Working directory:", Path.cwd())
````
