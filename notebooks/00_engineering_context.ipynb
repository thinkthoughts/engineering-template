## Load the repository in Colab or local Jupyter

This bootstrap cell:

- uses the current repository when it is already available;
- clones `engineering-template` into a fresh Colab workspace when needed;
- adds `src/` to Python's import path;
- changes the working directory to the repository root.

````python
from pathlib import Path
import os
import subprocess
import sys

REPOSITORY_URL = (
    "https://github.com/thinkthoughts/"
    "engineering-template.git"
)
PACKAGE_NAME = "engineering_template"
CURRENT = Path.cwd().resolve()

ROOT = next(
    (
        candidate
        for candidate in (
            CURRENT,
            CURRENT.parent,
        )
        if (
            candidate
            / "src"
            / PACKAGE_NAME
        ).exists()
    ),
    None,
)

if ROOT is None:
    ROOT = Path("/content/engineering-template")

    if not ROOT.exists():
        subprocess.run(
            [
                "git",
                "clone",
                REPOSITORY_URL,
                str(ROOT),
            ],
            check=True,
        )

SRC = ROOT / "src"
PACKAGE_DIR = SRC / PACKAGE_NAME

if not PACKAGE_DIR.exists():
    raise RuntimeError(
        f"Expected Python package at {PACKAGE_DIR}"
    )

if str(SRC) not in sys.path:
    sys.path.insert(
        0,
        str(SRC),
    )

os.chdir(ROOT)

print("Repository root:", ROOT)
print("Python source:", SRC)
print("Working directory:", Path.cwd())
````

## Specification Grammar

\[
\text{Constraint}
\rightarrow
\text{Connected lane}
\rightarrow
\text{Engineering object}
\rightarrow
\text{Engineering variable}
\rightarrow
\text{Observable state}
\rightarrow
\text{Indicator}
\]

**Admissible generalizations trail leading specifications.**

## Engineering Variable Specification

\[
C_t \rightarrow T_t \rightarrow x_t \rightarrow I_t
\]

- \(C_t\): constraint
- \(T_t\): telemetry
- \(x_t\): engineering variable
- \(I_t\): indicator

> Constraints specify telemetry. Telemetry specifies an engineering variable. The engineering variable specifies indicators.
