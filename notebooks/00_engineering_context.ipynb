# Engineering Context

**Repository:** `benefits-distribution`  
**Engineering statement:** **Benefit Distribution Specifies Sustainable Development**

> Leading constraints specify connected lanes.  
> Connected lanes specify engineering variables.  
> Observable states specify indicators.  
> Trailing indicators evaluate deployment.

Notebook 00 initializes the engineering context used by every later notebook in this repository.

## 1. Repository configuration

Only this cell is repository-specific.

For a future repository, change:

- the repository name,
- the engineering statement,
- the lane specification,
- the construction sequence,
- the source and next notebook.

```python
from dataclasses import dataclass, asdict
from pathlib import Path
from zipfile import ZipFile, ZIP_DEFLATED
import json
import numpy as np
import matplotlib.pyplot as plt

@dataclass(frozen=True)
class RepositoryContext:
    repository: str
    engineering_statement: str
    constraint: str
    lane_symbols: tuple[str, ...]
    lane_labels: tuple[str, ...]
    engineering_variables: tuple[str, ...]
    observable_states: tuple[str, ...]
    indicators: tuple[str, ...]
    construction_sequence: tuple[str, ...]
    next_notebook: str
    source: str = ""
    notes: str = ""

    def validate(self) -> None:
        scalar_fields = {
            "repository": self.repository,
            "engineering_statement": self.engineering_statement,
            "constraint": self.constraint,
            "next_notebook": self.next_notebook,
        }
        for name, value in scalar_fields.items():
            if not value.strip():
                raise ValueError(f"{name} must be non-empty")

        sequence_fields = {
            "lane_symbols": self.lane_symbols,
            "lane_labels": self.lane_labels,
            "engineering_variables": self.engineering_variables,
            "observable_states": self.observable_states,
            "indicators": self.indicators,
            "construction_sequence": self.construction_sequence,
        }
        for name, values in sequence_fields.items():
            if not values:
                raise ValueError(f"{name} must contain at least one item")
            if any(not str(item).strip() for item in values):
                raise ValueError(f"{name} contains an empty item")

        if len(self.lane_symbols) != len(self.lane_labels):
            raise ValueError("lane_symbols and lane_labels must have equal length")


CONTEXT = RepositoryContext(
    repository="benefits-distribution",
    engineering_statement=(
        "Benefit Distribution Specifies Sustainable Development"
    ),
    constraint="deployment constraints",
    lane_symbols=(
        r"$\pi_t$",
        r"$T_t$",
        r"$d_t$",
        r"$I_t$",
    ),
    lane_labels=(
        "policy",
        "telemetry",
        "distribution state",
        "development indicators",
    ),
    engineering_variables=(
        "deployment policy",
        "deployment telemetry",
        "benefit-distribution state",
        "development indicators",
    ),
    observable_states=(
        "deployment telemetry",
        "benefit-distribution state",
    ),
    indicators=(
        "sustainable-development indicators",
    ),
    construction_sequence=(
        "00 Engineering context",
        "07 Telemetry",
        "13 State estimation",
        "17 Distribution",
        "23 Dynamics",
        "29 Optimization",
        "37 Controllers",
        "43 Benchmarks",
    ),
    next_notebook="07_telemetry.ipynb",
    source=(
        "Google DeepMind: "
        "The Case for Globally Beneficial Technology"
    ),
    notes=(
        "Standalone Notebook 00 for the benefits-distribution repository."
    ),
)

CONTEXT.validate()
CONTEXT
```

## 2. Initialize output paths

This notebook is self-contained and writes only to:

- `figures/`
- `results/`
- `data/`

```python
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

FIGURES = ROOT / "figures"
RESULTS = ROOT / "results"
DATA = ROOT / "data"

for path in (FIGURES, RESULTS, DATA):
    path.mkdir(parents=True, exist_ok=True)

def savefig(filename: str) -> Path:
    path = FIGURES / filename
    plt.savefig(path, dpi=240, bbox_inches="tight")
    print("saved:", path)
    return path

print("ROOT:", ROOT)
```

## 3. Engineering Context

The shared engineering grammar is:

\[
\text{Constraint}
\rightarrow
\text{Connected lanes}
\rightarrow
\text{Engineering variables}
\rightarrow
\text{Observable states}
\rightarrow
\text{Indicators}
\]

This figure remains the same across repositories.

```python
# Figure 1 — Engineering Context

grammar = [
    "Constraint",
    "Connected lanes",
    "Engineering variables",
    "Observable states",
    "Indicators",
]

plt.figure(figsize=(8, 8.5))
ax = plt.gca()
ax.axis("off")

ys = np.linspace(0.82, 0.20, len(grammar))
x = 0.5

for index, (label, y) in enumerate(zip(grammar, ys)):
    width = 0.54
    ax.add_patch(
        plt.Rectangle(
            (x - width / 2, y - 0.045),
            width,
            0.09,
            fill=False,
            linewidth=1.8,
        )
    )
    ax.text(
        x,
        y,
        label,
        ha="center",
        va="center",
        fontsize=13,
        fontweight="bold" if index < 3 else None,
    )

    if index < len(grammar) - 1:
        ax.annotate(
            "",
            xy=(x, ys[index + 1] + 0.055),
            xytext=(x, y - 0.055),
            arrowprops=dict(
                arrowstyle="->",
                linewidth=1.8,
            ),
        )

ax.text(
    0.5,
    0.94,
    "Engineering Context",
    ha="center",
    fontsize=18,
    fontweight="bold",
)

ax.text(
    0.5,
    0.08,
    "Engineering begins by specifying variables before evaluating indicators.",
    ha="center",
    fontsize=10.5,
)

savefig("00_engineering_context.png")
plt.show()
```

## 4. Repository Lane Specification

This is the repository-specific figure.

For `benefits-distribution`:

\[
\pi_t
\rightarrow
T_t
\rightarrow
d_t
\rightarrow
I_t
\]

The symbols and labels come directly from the configuration cell.

```python
# Figure 2 — Repository Lane Specification

symbols = CONTEXT.lane_symbols
labels = CONTEXT.lane_labels

plt.figure(figsize=(12, 4.8))
ax = plt.gca()
ax.axis("off")

xs = np.linspace(0.12, 0.88, len(symbols))
y = 0.54

for index, (x, symbol, label) in enumerate(zip(xs, symbols, labels)):
    ax.add_patch(
        plt.Rectangle(
            (x - 0.075, y - 0.09),
            0.15,
            0.18,
            fill=False,
            linewidth=1.8,
        )
    )
    ax.text(
        x,
        y + 0.025,
        symbol,
        ha="center",
        va="center",
        fontsize=20,
    )
    ax.text(
        x,
        y - 0.055,
        label,
        ha="center",
        va="center",
        fontsize=9,
    )

    if index < len(symbols) - 1:
        ax.annotate(
            "",
            xy=(xs[index + 1] - 0.09, y),
            xytext=(x + 0.09, y),
            arrowprops=dict(
                arrowstyle="->",
                linewidth=1.8,
            ),
        )

ax.text(
    0.5,
    0.90,
    "Repository Lane Specification",
    ha="center",
    fontsize=18,
    fontweight="bold",
)

ax.text(
    0.5,
    0.16,
    "Connected deployment lanes transform policy into observable indicators "
    "through measurable distribution.",
    ha="center",
    fontsize=10.5,
)

savefig("00_repository_lane_specification.png")
plt.show()
```

## 5. Repository Construction Sequence

Each notebook develops one engineering variable within the connected repository lane.

```python
# Figure 3 — Repository Construction Sequence

sequence = CONTEXT.construction_sequence

plt.figure(figsize=(13, 4.6))
ax = plt.gca()
ax.axis("off")

xs = np.linspace(0.07, 0.93, len(sequence))
y = 0.53

for index, (x, item) in enumerate(zip(xs, sequence)):
    number, title = item.split(" ", 1)

    ax.add_patch(
        plt.Rectangle(
            (x - 0.055, y - 0.10),
            0.11,
            0.20,
            fill=False,
            linewidth=1.8,
        )
    )

    ax.text(
        x,
        y + 0.045,
        number,
        ha="center",
        va="center",
        fontsize=13,
        fontweight="bold",
    )

    ax.text(
        x,
        y - 0.035,
        title.replace(" ", "\n", 1),
        ha="center",
        va="center",
        fontsize=8.5,
    )

    if index < len(sequence) - 1:
        ax.annotate(
            "",
            xy=(xs[index + 1] - 0.065, y),
            xytext=(x + 0.065, y),
            arrowprops=dict(
                arrowstyle="->",
                linewidth=1.7,
            ),
        )

ax.text(
    0.5,
    0.90,
    "Repository Construction Sequence",
    ha="center",
    fontsize=18,
    fontweight="bold",
)

ax.text(
    0.5,
    0.18,
    "Each notebook develops one engineering variable "
    "within the connected repository lane.",
    ha="center",
    fontsize=10.5,
)

savefig("00_repository_construction_sequence.png")
plt.show()
```
